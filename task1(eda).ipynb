```python
import pandas as pd
df = pd.read_excel("School data.xlsx")
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum().sum()
df.duplicated().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['%_Math_Score'])
plt.show()
```

```python
sns.scatterplot(
    x='Parent_Literacy_Rate',
    y='%_Language_Score',
    data=df,
    s=4.5
)
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, fmt=".2f", annot_kws={"size": 6})
plt.show()
```

```python
Key insight:
1.Most students have average Math scores, with fewer students at very high
or very low levels.
2.There is a positive relationship between parent literacy and language scores,
meaning students with educated parents tend to perform better.
3.Math and Language scores are positively correlated, so students who do well in one
subject usually do well in others too.
4.The dataset initially had missing values and duplicates, but after cleaning, it became
more accurate and reliable.
5.Not all factors strongly affect performance — the heatmap shows some weak
relationships, meaning certain variables have less impact.
```
